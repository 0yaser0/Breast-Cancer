### Ce script permet de charger un fichier CSV, puis de remplacer les valeurs spécifiques `9` par `NaN` dans un ensemble de colonnes données


```python


columns_to_replace = ['race_eth', 'first_degree_hx', 'age_menarche', 'age_first_birth', 
                      'BIRADS_breast_density', 'current_hrt', 'menopaus', 
                      'bmi_group', 'biophx', 'breast_cancer_history']


data[columns_to_replace] = data[columns_to_replace].replace(9, np.nan)
```

### calcul la frequance 

```python
missing_percentage = (data.isnull().sum() / len(data)) * 100
missing_percentage

```

```python
data.corr()
```

### affichier la correlation entre les colones 


```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Matrice de Corrélation')
plt.show()
```

### Corrélation faible entre les variables : D'après la matrice de corrélation, on remarque qu'il n'y a pas de forte corrélation linéaire entre les variables. Cela signifie qu'il n'existe pas de relation directe évidente entre les différentes variables. De plus, étant donné que certaines de ces variables sont catégorielles, nous ne pouvons pas modifier ou remplacer ces données de manière directe.
